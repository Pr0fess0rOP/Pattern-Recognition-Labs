# Question 3: Mean, Deviation, Variance, Max and Min

```python
def meanFunction(x):
    return sum(x)/len(x)
```

```python
def stdFunction(x):
    return varianceFunction(x)**(0.5)
```

```python
def varianceFunction(x):
    mu = meanFunction(x)
    var = [(i-mu)**2 for i in x]
    return sum(var)/(len(var)-1)
```

```python
def maxFunction(x):
    return max(x)
```

```python
def minFunction(x):
    return min(x)
```

```python
col = df.columns

for feature in col:
    print(f'Column: {feature}')
    print("Mean = ",meanFunction(df[feature]))
    print("Standard Deviation = ",stdFunction(df[feature]))
    print("Variance = ",varianceFunction(df[feature]))
    print("Max = ",minFunction(df[feature]))
    print("Min = ",maxFunction(df[feature]))
    print("\n")
```

# Question 5: Standardization and Normalization

```python
df_normalized = df

def normalize(col,mini,maxi):
    setdiff = maxi-mini
    return (col-mini)/setdiff

dict_col_values = {}
for i in df.columns:
    dict_col_values[i] = [min(df_normalized[i]),max(df_normalized[i])]
    
for i in df.columns:
    mini,maxi = dict_col_values[i][0],dict_col_values[i][1]
    df_normalized[i] = df_normalized[i].apply(normalize,args = (mini,maxi))

df.describe()
```
