```python
import numpy as np
import matplotlib.pyplot as plt
from math import floor, ceil

linalg = np.linalg

N = 1000 #given
p = [0.6,0.3,0.1] #given
l = 3 #given

mu = [np.transpose([1,1]),np.transpose([7,7]),np.transpose([15,1])] #given

cov = np.identity(2) #given
```

# Question 1

```python
for i in range(l):
    # x, y = np.random.multivariate_noraml( cov, mu[i], ceil(N/1) )
    x, y = np.random.multivariate_normal( mu[i], cov, floor(N/l) ).T
    plt.scatter(x,y)
    
plt.show()
```

# Question 2

```python
cov = [ np.array([[12,0],[0,1]]), np.array([[8,3],[3,2]]), np.array([[2,0],[0,2]]) ] #given

for i in range(l):
    # x, y = np.random.multivariate_noraml( cov, mu[i], ceil(N/1) )
    x, y = np.random.multivariate_normal( mu[i], cov[i], floor(N/l) ).T
    plt.scatter(x,y)
    
plt.show()
```

# Question 3

```python
#same as question 1
cov = np.identity(2) #given

for i in range(l):
    # x, y = np.random.multivariate_noraml( cov, mu[i], int(N*(p[i])) )
    x, y = np.random.multivariate_normal( mu[i], cov, int(N*p[i]) ).T
    plt.scatter(x,y)
    
print("Part (1) where covariance is I(2): ")
plt.show()
```
