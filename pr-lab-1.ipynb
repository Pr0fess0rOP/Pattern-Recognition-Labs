# Question 1: Generate Random Number

```python
def randomGenerator(seed, modular, multiplier, n):
    randomList = []
    randomList.append(seed)
    while(n):
        randomList.append((randomList[-1]*multiplier)%modular)
        n= n-1
    print(randomList)

randomGenerator(2,6,2,20)
```

# Question 2: Generate Random Number

```python
def randomGenerator2(seed, modular, multiplier, const, n):
    randomList = []
    randomList.append(seed)
    while(n):
        randomList.append(((randomList[-1]*multiplier)+const)%modular)
        n= n-1
    print(randomList)

randomGenerator2(2,6,2,5,20)
```

# Question 3: Generate first 15 values

```python
def first15Generator(n, boundaryList):
    for i in range(n):
        boundaryList.append((3*boundaryList[-1] + 5*boundaryList[-2])%100)
    return boundaryList

first15Generator(15, [23,66])
```

# Question 4: Approximate

```python
from numpy import corrcoef
from random import uniform
from math import sqrt
import random as rd

def mean(lst):
    return sum(lst)/len(lst)

def corrFunc(u,v):
    n = len(u)
    uv = []
    u2 = []
    v2 = []
    for i in range(n):
        uv.append(u[i]*v[i])
        u2.append(u[i]*u[i])
        v2.append(v[i]*v[i])
    return ((n*sum(uv))-(sum(u)*sum(v)))/((sqrt(n*sum(u2)-sum(u)*sum(u)))*(sqrt(n*sum(v2)-sum(v)*sum(v))))

u = [] 
va = [] 
vb = [] 
n = 100 

for i in range(n):
    u.append(rd.uniform(0,1))
    va.append(sqrt(1 - (u[-1]*u[-1])))
    vb.append(u[-1]*u[-1])
    
answerA = corrcoef(u,va)
answerB = corrcoef(vb,va)

print(answerA)
print(corrFunc(u,va))
print(answerB)
print(corrFunc(vb,va))
```

# Question 5:

```python
from random import uniform
import matplotlib.pyplot as plt

sample_size = [50,100,150,200,250,300,350,400,450,500]

def calculator(n):
    lengths = []
    while (n>0) :
        arr = []
        n = n-1
        while(sum(arr) <= 1):
            arr.append(uniform(0,1))
        lengths.append(len(arr))
    print(mean(lengths))
    plt.hist(lengths,bins=[0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10])
    plt.show()

for size in sample_size:
    calculator(size)
```

# Question 6:

```python
from math import exp
from random import random

def stimulator():
    counter = 0
    e = exp(-3)
    value = 1
    while(value>e):
        value = value*random()
        counter = counter+1
    return counter

sampleSpace = 100

N = [stimulator() for i in range(sampleSpace)]

plt.hist(N)

print(mean(N))

for i in range(7):
    print("P(N={}) = ".format(i),N.count(i)/sampleSpace)
```

# Question 7: Generate Scaler Normal Random Numbers and Plot 

```python
import numpy as np
import matplotlib.pyplot as plt

def calculator(myu,sigma,n):
    return(np.random.normal(myu,sigma,n))

myuSample = [0.3,0.5,0.7]
sigmaSample = [0.1,0.2,0.3]
sampleSize = [50,100,150,200,250,300,350,400,450,500]
N = []

for n in sampleSize:
    for myu in myuSample:
        for sigma in sigmaSample:
            N.append(calculator(myu,sigma,n))
            plt.hist(N[-1])
            print(f'myu = {myu}, sigma = {sigma}, n = {n}')
            plt.show()
```
