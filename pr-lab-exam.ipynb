# Question (a)

```python
def roll():
    dice1 = random.choices([1,2,3,4,5,6], k = 1)[0]
    dice2 = random.choices([1,2,3,4,5,6], k = 1)[0]
    return dice1, dice2
iterationList = []    
def driverCode():
    obtainedSet = set()
    # doneList = [False for i in range(11)]
    counter = 0

    while(sum(obtainedSet) != 77):
        dice1, dice2 = roll()
        summation = dice1 + dice2
        obtainedSet.add(summation)
        counter += 1
    iterationList.append(counter)
    
answer = []
for i in range(100):
    driverCode()
    answer.append(sum(iterationList)/len(iterationList))
print(sum(answer)/len(answer))
print(answer)
```
